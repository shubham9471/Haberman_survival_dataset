# CDF

```python
class_1 = Haberman.loc[Haberman["Survival_status"] == 1] #1 means the patient will survive 
class_2 = Haberman.loc[Haberman["Survival_status"] == 2] #2 means the patient will die
```

```python
#AGE
#Survive
count1, bin_edges1 = np.histogram(class_1['Age'], bins=10, density = True)
pdf1=count1/sum(count1)
cdf1=np.cumsum(pdf1)

#Die
count2, bin_edges2 = np.histogram(class_2['Age'], bins=10, density = True)
pdf2=count2/sum(count2)
cdf2=np.cumsum(pdf2)
sns.set_style('darkgrid')
plt.plot(bin_edges1[1:], pdf1, label = 'pdf1')
plt.plot(bin_edges1[1:], cdf1, label = 'cdf1')
plt.plot(bin_edges1[1:], pdf2, label = 'pdf2')
plt.plot(bin_edges1[1:], cdf2, label = 'cdf2')
plt.xlabel('Age')
plt.title('(CDF PLOT 1.1) AGE CORRESPONDING TO SURVIVAL AND DEATH RATE ')
plt.legend()
plt.show()
```

# Observations ;-
1. Both the grids are overlapping
2. At the age of 45 the survival rate is higher.

```python
#Year_of_Operation
#Survive
count1, bin_edges1 = np.histogram(class_1['year_of_operation'], bins=10, density = True)
pdf1=count1/sum(count1)
cdf1=np.cumsum(pdf1)

#Die
count2, bin_edges2 = np.histogram(class_2['year_of_operation'], bins=10, density = True)
pdf2=count2/sum(count2)
cdf2=np.cumsum(pdf2)
sns.set_style('darkgrid')
plt.plot(bin_edges1[1:], pdf1, label = 'pdf1')
plt.plot(bin_edges1[1:], cdf1, label = 'cdf1')
plt.plot(bin_edges1[1:], pdf2, label = 'pdf2')
plt.plot(bin_edges1[1:], cdf2, label = 'cdf2')
plt.xlabel('year_of_operation')
plt.title('(CDF PLOT 1.2) Year_of_Operation CORRESPONDING TO SURVIVAL AND DEATH RATE ')
plt.legend()
plt.show()
```

# Observations:-
1.Again Both the graphs of  pdf and cdf are overlapping
2. between the age of 60 to 64  chances of survival is higher than that of  66 to 70. 

```python
#Positive Auxillary
#Survive
count1, bin_edges1 = np.histogram(class_1['Positive_auxillary'], bins=10, density = True)
pdf1=count1/sum(count1)
cdf1=np.cumsum(pdf1)

#Die
count2, bin_edges2 = np.histogram(class_2['Positive_auxillary'], bins=10, density = True)
pdf2=count2/sum(count2)
cdf2=np.cumsum(pdf2)
sns.set_style('darkgrid')
plt.plot(bin_edges1[1:], pdf1, label = 'pdf1')
plt.plot(bin_edges1[1:], cdf1, label = 'cdf1')
plt.plot(bin_edges1[1:], pdf2, label = 'pdf2')
plt.plot(bin_edges1[1:], cdf2, label = 'cdf2')
plt.xlabel('Positive_auxillary')
plt.title('(CDF PLOT 1.3) Positive Auxillary CORRESPONDING TO SURVIVAL AND DEATH RATE ')
plt.legend()
plt.show()
```

# Observations:
1. Both the graphs are overlapping
2. Somewhere between 7 to 23 postitve auxillary the survival rate is higher than death rate.

# CONTOUR PLOT

```python
sns.jointplot(data = Haberman, x = "Age", y = "year_of_operation", kind = "kde")
plt.plot(xlabel = "Age")
plt.plot(ylabel = "year_of_operation")
plt.suptitle("CONTOUR PLOT OF AGE VS YEAR OF OPERATION")
plt.show()
```

```python
sns.jointplot(data = Haberman, x = "Age", y = "Positive_auxillary", kind = "kde")
plt.plot(xlabel = "Age")
plt.plot(ylabel = "Positive_auxillary")
plt.suptitle("CONTOUR PLOT OF AGE VS POSTIVE AUXILLARY")
plt.show()
```
